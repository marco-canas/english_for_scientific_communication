### **Metodolog√≠a de Ense√±anza: Aprendizaje Basado en Problemas y Proyectos (ABPP)**



El **Aprendizaje Basado en Problemas y Proyectos (ABPP)** es una metodolog√≠a educativa que integra los enfoques del **Aprendizaje Basado en Problemas (ABP)** y el **Aprendizaje Basado en Proyectos (PBL, por sus siglas en ingl√©s)**. 

Es especialmente √∫til para la formaci√≥n de futuros docentes, ingenieros y administradores, ya que combina el desarrollo de habilidades pr√°cticas con el pensamiento cr√≠tico y reflexivo. 

A continuaci√≥n, se presenta una descripci√≥n adaptada a la ense√±anza de las matem√°ticas en educaci√≥n infantil, Licenciatura en Matem√°ticas, Ingenier√≠a Agropecuaria y Administraci√≥n de Empresas:

---



### **1. Fundamentos del ABPP**


El ABPP se basa en:
- **Resoluci√≥n de Problemas Reales:** Los estudiantes analizan y resuelven problemas aut√©nticos relacionados con su √°rea de formaci√≥n, fomentando la aplicaci√≥n pr√°ctica del conocimiento.


- **Proyectos como Herramientas Integradoras:** Los proyectos permiten desarrollar habilidades interdisciplinarias y ofrecer soluciones concretas a los problemas planteados.


- **Aprendizaje Activo:** Los estudiantes son responsables de su aprendizaje y participan activamente en el proceso de investigaci√≥n, dise√±o y soluci√≥n.

---

## **1. Elementos de gram√°tica (Present Simple en contexto matem√°tico)**



En el contexto matem√°tico, el presente simple se usa para expresar **hechos universales, propiedades y definiciones**.



**Afirmativa:**



* **Estructura:** `Sujeto + verbo en presente simple + complemento`

  * *Two plus two equals four.*
  * *A triangle has three sides.*
  * *The sum of angles in a triangle is 180 degrees.*



**Negativa:**

* **Estructura:** `Sujeto + do/does + not + verbo base + complemento`

  * *A square does not have three sides.*
  * *Five plus five does not equal eleven.*



**Interrogativa:**

* **Estructura:** `Do/Does + sujeto + verbo base + complemento?`

  * *Does a pentagon have five sides?* ‚Äì *Yes, it does.* / *No, it doesn‚Äôt.*
  * *Do two plus two equal five?* ‚Äì *No, they don‚Äôt.*

**Notas importantes:**

* Para tercera persona singular (he, she, it, nombres en singular), se a√±ade **-s** o **-es** al verbo (*equals*, *has*).
* En matem√°ticas, se usa el presente simple porque las afirmaciones son verdades atemporales.

- **Traducci√≥n de problemas matem√°ticos:** 

````python
 
equation = "x + 3 = 10"
numbers = [int(s) for s in equation.split() if s.isdigit()]
words = [p.number_to_words(num) for num in numbers]
print(words)  # Output: ['three', 'ten']
  
````


- **Generar enunciados en ingl√©s:** 

````python
 

def translate_problem(problem):
    parts = problem.split()
    for i, part in enumerate(parts):
        if part.isdigit():
            parts[i] = p.number_to_words(int(part))
    return ' '.join(parts)

print(translate_problem("Solve 5 + 7"))  
# Output: "Solve five + seven"



````

### **üìö Ejemplo Completo para Matem√°ticas**


````python

import inflect
import re  # Importamos el m√≥dulo de expresiones regulares

p = inflect.engine()

def explain_problem(problem):
    # Usamos regex para encontrar todos los n√∫meros (enteros o decimales)
    number_pattern = r"\d+\.?\d*"  # Detecta n√∫meros como 2, 15, 3.14, etc.
    numbers = re.findall(number_pattern, problem)
    words = [p.number_to_words(int(float(num))) for num in numbers]  # Convertimos a entero si es decimal
    
    # Reemplazar n√∫meros en el problema (conservando s√≠mbolos adyacentes como 'x' o ',')
    def replace_match(match):
        num = match.group()
        return p.number_to_words(int(float(num)))  # Mismo tratamiento que arriba
    
    problem_in_words = re.sub(number_pattern, replace_match, problem)
    
    print("Problema en palabras:", problem_in_words)
    print("N√∫meros extra√≠dos:", words)

explain_problem("If 2x + 3 = 15, find x")
````

**Output:**  
```
Problema en palabras: If twox + three = fifteen, find x
N√∫meros extra√≠dos: ['two', 'three', 'fifteen']
```
